### Rolling with Advantage

Simulate rolling a 20-sided die, where one rolls the die twice, and takes the maximum of the two rolls. Plot a histogram of the results of doing this 1,000,000 times. Does the shape of the histogram make sense?

Results: The histogram shows the probability increasing linearly as the value on the die increases. This makes sense since if you roll a high value for the first roll, e.g. 16, it is less likely that your second roll will give you a value higher than that, than if you were to roll a low value, e.g. 6, for the first roll. A high number like 16 only has four values which are higher than it (17, 18, 19, 20), while a low value like 6 has fourteen values which are higher than it (7, 8, 9, ... 19, 20), meaning it is more likely that the 6, if rolled first, will be replaced by a higher value with the second roll.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def roll_some_dice(num_sides, num_rolls):
    """
    This function rolls a die with a given number of sides, a given number of times.
    It calculates the average roll value and creates a histogram plot of all the rolls.
    
    :param num_sides: number of sides on the die being rolled
    :param num_rolls: number of times the die is being rolled
    :return: average roll value and histogram of all rolls
    """
    
    total = []
    for i in range(num_rolls):
        total.append(np.random.randint(1, num_sides+1))
        # print(np.random.randint(1, num_sides+1))
    
    bins = [(i+0.5) for i in range(num_sides+1)]
    plt.hist(total, bins=bins, rwidth=0.8)
    plt.xlabel('Value Rolled')
    plt.ylabel('Number of Rolls')
    plt.xticks([(i+1) for i in range(num_sides)])
    # print(total)

    return sum(total)/num_rolls
```

```python
sides = 20
rolls = 1000000

avg = roll_some_dice(sides, rolls)
print("Average Roll Value: ", avg)
print("Theoretical Average Roll Value: ", sum([(i+1) for i in range(sides)])/sides)
print("num_rolls/num_sides: ", rolls/sides)
```

```python
def max_value_rolled(n, m):
    """
    This function rolls a die with n sides, a total of m times,
    and returns the maximum value rolled.
    
    :param n: number of sides on the die
    :param m: number of times the die is rolled
    :return: the maximum value rolled
    """
    
    max_val = 1
    for i in range(m):
        current = np.random.randint(1, n+1)
        if current > max_val:
            max_val = current
    
    return max_val
```

```python
num_times = 1000000

rolling_with_advantage = []
for i in range(num_times):
    rolling_with_advantage.append(max_value_rolled(20, 2))

bins = [(i+0.5) for i in range(21)]
plt.hist(rolling_with_advantage, bins=bins, rwidth=0.8)
plt.xlabel('Value Rolled')
plt.ylabel('Number of Rolls')
plt.xticks([(i+1) for i in range(20)])
```
